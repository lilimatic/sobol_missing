# Options

#### The price of an option is defined as 

\begin{equation}
C_0 =e^{-r T} \mathbb{E}_{\mathbb{Q}} \left[ \max\{\left(S_T-K\right),0\}\right],
\end{equation}
#### where $T$ is the maturity, $K$ the strike price, $r$ is the risk-free rateand  $\mathbb{Q}$ is the risk neutral measure with the density $q \left( \cdot \right)$. Options give us the right, but not the obligation, to buy or sell an asset $S_t, t \in [0,T]$ in the future at time $T$. 

#### A vanilla call option $C_{t}(K,T)$ gives us the right to purchase an asset at time $T$.  The payoff of a call option is 
$$\max \left\{ S_{T}-K,0 \right\},$$
#### where $K$ denotes the strike price. The intrinsic value of this option is 
\begin{equation}
\max \left\{ S_{t}-K,0 \right\}.
\end{equation}
#### at time $t \in (0,T ]$.

#### A vanilla put option $P_{t}(K,T)$ gives us the right sell the asset at time $T$.  The payoff of a call option is 
$$\max \left\{ K-S_{T},0 \right\},$$
#### for strike price $K$. The intrinsic value of this option is 
\begin{equation}
\max \left\{ K-S_{t},0 \right\}
\end{equation}
#### at time $t \in (0,T ]$.

#### The relationship between calls and puts with the same strike is given by the put call parity
\begin{equation}
C(K,T)-P(K,T)=S_0-K e^{-r T}.
\end{equation}

#### The following illustrations show the payoff of a call and put, respectively. 

```python
call = [max(x-100,0)-20 for x in np.linspace(0,200,9)]

put  = [max(100-x,0)-20 for x in np.linspace(0,200,9)] 

plt.plot(np.linspace(0,200,9),call)
plt.xlabel("Undlying")
plt.ylabel("Payoff")
plt.title("Payoff of a call")

plt.show()

plt.plot(np.linspace(0,200,9),put,color='red')
plt.xlabel("Underlying")
plt.ylabel("Payoff")
plt.title("Payoff of a put")
```

## Moneyness

#### The relationship between the underlying $S_t, t \in [0,T]$ and the strike is referred to as the moneyness of an option. An option is in

#### 1. At-The-Money (ATM), if the strike price $ K = S_t $ is equal to the price of the underlying. 

#### 2. In-The-Money (ITM), if the strike price $  K < S_t  $ is below the price of the underlying. 

#### 3. Out-Of-The-Money (OTM), if the strike price $  K > S_t  $ is above the price of the underlying.

# Option valuation using the Fourier Transform (Carr and Madan)

- ## Landmark paper:  Carr, Peter; Madan, Dilip B. (1999). "Option valuation using the fast Fourier transform". Journal of Computational Finance. 2 (4): 6173

### The following option pricing method is an elegant way to compute option prices. 

- ### Key idea: Use the characteristic function of the log-asset $ln(S_t)$ to compute an option price

## Concept and derivation 

#### This section derives the CarrMadan formula. According to the concepts of arbitrage theory the price of an option is 

\begin{equation}
\begin{aligned}
C_0 &=e^{-r T} \mathbb{E}_{\mathbb{Q}} \left[ \max\{\left(S_T-K\right),0\}\right] = e^{-r T} \int_k^{\infty}\left(e^s-e^k\right) q_T(s) d s,
\end{aligned}
\end{equation}
#### where $q_T(s)$ is the density of the log-price of the underlying $s = ln(S_T)$ and $k = ln(K)$. Let 
\begin{equation}
\phi_T(t) \equiv \int_{-\infty}^{\infty} e^{\mathrm{i} t s} q_T(s) \mathrm{d} s
\end{equation}
#### be the characteristic function of this density $q_T(\cdot)$. We consider a damped option price
\begin{equation}
c_T(k)=e^{\alpha k} C_T(k)
\end{equation}
#### that Carr and Madan (1999) scale by this damping factor $\alpha>0$ to ensure the existence of the Fourier transform. The Fourier transform of $c_T(k)$ is 
\begin{equation}
\psi_T(t)=\int_{-\infty}^{\infty} e^{\mathrm{i} t k} c_T(k) \mathrm{d} k. 
\end{equation}
#### Accordingly, the option price is equal to
\begin{equation}
C_T(k)=\frac{\exp (-\alpha k)}{2 \pi} \int_{-\infty}^{\infty} \mathrm{e}^{-\mathrm{i} t k} \psi_T(t) d t=\frac{\exp (-\alpha k)}{\pi} \int_0^{\infty} \mathrm{e}^{-\mathrm{i} t k} \psi(t) d t.
\end{equation}
#### This means that the option price can be obtained when we have an analytic expression of the characteristic function $\psi_T(\cdot)$. The following is
\begin{equation}
\begin{aligned}
\psi_T(t) & =\frac{\mathrm{e}^{-r T} \phi_T(t-(\alpha+1) \mathrm{i})}{\alpha^2+\alpha-t^2+\mathrm{i}(2 \alpha+1) t},
\end{aligned}
\end{equation}
#### which you will need to show in exercise 2.2. 

#### The challenge is to numerically approximate the integral $\int_0^{\infty} \mathrm{e}^{-\mathrm{i} t k} \psi(t) d t$.

# Black-Scholes Option pricing

The Black-Scholes option pricing model assumes that the log-returns of options are log-normal.


The dynamics under the risk-neutral measure are 

\begin{equation}
d S_t=r S_t d t+\sigma S_t d W_t,
\end{equation}


with the solution 

\begin{equation}
S_t=S_0 \exp \left\{\left(r-\frac{\sigma^2}{2}\right) t+\sigma W_t\right\}.
\end{equation}


```python
def asset_rn(S0,r,sigma,M,N):
    # Function
    # - S0    : Initial price of the underlying
    # - r     : Risk free interest rate
    # - sigma : Volatility
    # - M     : Number of days 
    # - N     : Number of simulations
    St    = np.zeros([N , M +1])
    Z     = np.random.normal(size =[N , M ])
    St[:,0]          = S0
    for i in range (M):
        St[: , i +1] = St [: , i ]* np . exp(( r - sigma **2/2)* dt+ sigma * np.sqrt( dt )* Z[: , i ])
    return St
```

```python
npr.seed(123)
M     = 365
N     = 10
S0    = 100
K     = 100
T     = 1
dt    = T/M
mu    = 0.3
sigma = 0.2
r     = 0.1


St = asset_rn(S0,r , sigma,M,N)

plt.plot(np.transpose(St))
plt.ylabel('Asset price process')
plt.xlabel('time steps')
plt.show()
```

# Black-Scholes Option Pricing Model

Black and Scholes (1976) introduce the Black-Scholes-Merton Model for option pricing 

\begin{equation}
C\left(S_t, t\right)=S_t \cdot \Phi\left(d_1\right) -K \cdot \Phi\left(d_2\right) e^{-r(T-t)},
\end{equation}
where 

\begin{equation}
\begin{array}{l}
d_1=\frac{1}{\sigma \sqrt{T-t}}\left\{\ln \left(\frac{S_t}{K}\right)+\left(r+\frac{\sigma^2}{2}\right)(T-t)\right\} \\
d_2=\frac{1}{\sigma \sqrt{T-t}}\left\{\ln \left(\frac{S_t}{K}\right)+\left(r-\frac{\sigma^2}{2}\right)(T-t)\right\}.
\end{array}
\end{equation}

The term $\Phi\left(\cdot\right)$ is the cumulative distribution function of the standard normal distribution.

The Delta of this option is 
\begin{equation}
\Delta=\frac{\partial C}{\partial S} = \Phi\left(d_1\right).
\end{equation} 

```python
#From other CUATS Session

from scipy.stats import norm

Phi = norm.cdf

def bs(sigma, T,t, K, St, r):
    d_1 = 1/sigma * 1/np.sqrt(T-t) * (np.log(St/K) + (r + sigma**2/2) * (T-t))
    d_2 = d_1 - sigma * np.sqrt(T-t)
    term_1 = St * Phi(d_1)
    term_2 = K * np.exp(-r * (T-t)) * Phi(d_2)
    return term_1 - term_2


print('Longer-dated options are more expensive, else there is potential for (calender) arbitrage.')
print(bs(0.2, 30,0, 100, 100, 0.2))

print(bs(0.2, 60,0, 100, 100, 0.2))

def delta_bs(sigma, T,t, K, St, r):
    d_1 = 1/sigma * 1/np.sqrt(T-t) * (np.log(St/K) + (r + sigma**2/2) * (T-t))
    return Phi(d_1)
```

# Heston Model

#### Heston(1993) defines the Heston model as an extension of the Black-Scholes option pricing model. Loosely speaking, the model assumes that volatility is stochastic. 

#### Under the risk-neutral measure $\mathbb{Q}$, the dynamics of the Heston model are 
\begin{equation}
\begin{array}{c}
d S_t=r S_t d t+\sqrt{V_t} S_t d W_{t}^{s} \\
d V_t=\kappa \left(\theta-V_t\right) d t+\sigma_v \sqrt{V_t} d W_{t}^{v},
\end{array}
\end{equation}
#### where $W_{t}^{s}$ and $W_{t}^{v}$ are two correlated Wiener processes with correlation $\rho $ and $r$ is the risk-free rate. 

#### This model assumes that the volatility process $\sqrt{V_t}$ follows an Ornstein-Uhlenbeck process. Accordingly, the instantanous variance $V_t$ is described by a CoxIngersollRoss model
\begin{equation}
d V_t=\kappa \left(\theta-V_t\right) d t+\sigma_v \sqrt{V_t} d W_{t}^{v}
\end{equation}

#### The parameters of this stochastic volatility (SV) model are $\Theta_{SV} = \{\sigma_0, \kappa,\theta,\sigma_v,\rho\}$, where

- #### $\sigma_0$ is the initial variance level $V(0) = \sigma_0$
- #### $\theta$ is the long-term price variance, or mean-reverse level. 
- #### $\kappa$ is the reversion rate to the long-term variance, or mean-reversion rate. This is the rate at which the instantanous variance $V_t$ reverts to the long-term variance $\theta$
- #### $\sigma_v$ is volatility-of-volatility (vol-of-vol) parameter, that provides the variance of $V_t$
- #### $\rho$ is the correlation between the two Wiener processes $W_{t}^{s}$ and $W_{t}^{v}$. It indicates the correlation between an asset and it's volatility. In most markets, we observe a "leverage-effect": When prices go down, volatility increases. Hence, $\rho$ is typically negative. 

#### As variance should not be negative, the parameters need to satisfy the Feller condition
\begin{equation}
2 \kappa \theta> \sigma_v^{2},
\end{equation}
#### such that $V_{t}$ is strictly positive.

# Computing greeks 

### There are various possibilies to compute greeks. The Black-Scholes-Merton model, as well as many other affine-jump diffusion models, have analytic solutions of market sensitivities. If there is no analytic solution to compute the price of an option, greeks can be computed numerically. 

### We present the greeks first-order greeks computation with finite differences


\begin{array}{l}
\frac{\partial C}{\partial S} \approx \frac{C(S+\delta S, T, \sigma, r, K)-C(S, T, \sigma, r, K)}{\Delta S} \\
\frac{\partial^2 C}{\partial S^2} \approx \frac{C(S+\delta S, T, \sigma, r, K)-2 C(S, T, \sigma, r, K)+C(S-\delta S, T, \sigma, r, K)}{(\delta S)^2} \\
\frac{\partial C}{\partial \sigma} \approx \frac{C(S, T, \sigma +\delta \sigma, r, K)-C(S, T, \sigma, r, K)}{\delta S} \\
\end{array}




```python
def delta_FD(sigma, T,t, K, St, r,dS):
    return (bs(sigma, T,t, K, St+dS, r) - bs(sigma, T,t, K, St, r))/dS

def gamma_FD(sigma, T,t, K, St, r,dS):
    return (bs(sigma, T,t, K, St+dS, r) - 2+ bs(sigma, T,t, K, St, r) + bs(sigma, T,t, K, St-dS, r) )/(dS)**2

def vega_FD(sigma, T,t, K, St, r,dvol):
    return (bs(sigma+dvol, T,t, K, St+dS, r) - bs(sigma, T,t, K, St, r))/dvol
```
